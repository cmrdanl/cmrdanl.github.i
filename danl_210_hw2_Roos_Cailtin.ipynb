```python
# Question 15
#Calculate the mean value of pass for the BUF posteam when all the following conditions hold:
    #wp is greater than 20% and less than 75%;
    #down is less than or equal to 2; and
    #half_seconds_remaining is greater than 120.
_spec = NFL2022_stuffs.query(" wp < .75 & wp > .20 & down <= 2 & half_seconds_remaining > 120 ")

_spec[_spec['posteam'] == 'BUF']['pass'].mean()
```

```python
# Question 17

keeping = ["Name", "Date", "Dumpster",'PlasticBottles', 'Polystyrene','CigaretteButts','GlassBottles', 'PlasticBags',"Wrappers", 'SportsBalls']


trashwheel_long = trashwheel[keeping].melt(id_vars = ["Name", 'Date', 'Dumpster'],
                                  var_name= 'Trash_Type',
                                  value_name="Number")


trashwheel_long.sort_values(["Name","Date","Dumpster","Trash_Type"], ascending= (True,True,True,True))
```
